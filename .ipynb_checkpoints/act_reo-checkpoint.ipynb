```python
twitter_array = []
with open('tweet-json.txt','r', encoding='utf-8') as twitter_file:
   
    # line serves as each object in the file
    for line in twitter_file:
        file = json.loads(line)
        tweet_id = file['id']
        retweet = file['retweet_count']
        likes = file['favorite_count']
                
        twitter_array.append({'tweet_id' : int(tweet_id),
                            'retweet' : int(retweet),
                            'likes' : int(likes),
                             })
        twitter_df = pd.DataFrame(twitter_array, columns=['tweet_id', 'retweet', 'likes'])
        
twitter_file = twitter_df.copy()
twitter_file.head(10)
```
