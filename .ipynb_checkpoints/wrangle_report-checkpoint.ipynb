### Data 2:
**Data Gathering**


**we've been given a json file which contains more details on the dataset we're working with and we've been asked to read the json  file into a dataframe.
so this contains the process of how we read the file.
first we make an empty array then open it as twitter_file ,we iterate each line in the twitter_file also naming the column needed .
we append the columns in the json file to the empty array named twitter_array.
lastly we read the twitter_array into a dataframe and get a preview.**

```python
twitter_array = []
with open('tweet-json.txt','r', encoding='utf-8') as twitter_file:
   
    # line serves as each object in the file
    for line in twitter_file:
        file = json.loads(line)
        tweet_id = file['id']
        retweet = file['retweet_count']
        likes = file['favorite_count']
      # add the column to the new array          
        twitter_array.append({'tweet_id' : int(tweet_id),
                            'retweet' : int(retweet),
                            'likes' : int(likes),
                             })
        # load the array into a new datafrme called twitter_df
        twitter_df = pd.DataFrame(twitter_array, columns=['tweet_id', 'retweet', 'likes'])
        
twitter_file = twitter_df.copy()
twitter_file.head(10)
```

**2. Filter out the columns that wont be necessary for the wrangling phase and we get a preview of what we're working with.**

```python
#split timestamp into three new columns 
df[['date','time','gmt']] = df['timestamp'].str.split(' ',n=2, expand=True)
df.head(3)
```

* **Get the mean of the column and replace the empty cells with it**

```python
df['in_reply_to_status_id'] = df['in_reply_to_status_id'].fillna(df['in_reply_to_status_id'].mean())
```

```python
df['in_reply_to_user_id'] = df['in_reply_to_user_id'].fillna(df['in_reply_to_user_id'].mean())
df.head(2)
```

```python
#replace the null values with the mean of the column
df['expanded_urls'].fillna(value=' ', inplace=True)
```

* **Drop columns that wont be neccesary further on**

```python
df.drop(['retweeted_status_id','retweeted_status_user_id','retweeted_status_timestamp','gmt'], axis=1, inplace=True)
df.shape
```

### Data Storing

* **Merge the new dataframes into 1 dataframe**

```python
#join all files into 1 masterset
all_file = [new_file1,new_file2,image_file]
all_file = [frame.set_index('tweet_id') for frame in all_file]
all_file = pd.concat(all_file,join='outer', axis=1)
all_file.shape
allfiles = all_file.copy()
```

* **Remove those rows that are retweets**

```python
# Remove retweets
regex = '^RT'
allfiles = allfiles[~allfiles['text'].str.match(regex, na=False)]
allfiles.shape
```
