```python
twitter_array = []
with open('tweet-json.txt','r', encoding='utf-8') as twitter_file:
   
    # line serves as each object in the file
    for line in twitter_file:
        file = json.loads(line)
        tweet_id = file['id']
        retweet = file['retweet_count']
        likes = file['favorite_count']
                
        twitter_array.append({'tweet_id' : int(tweet_id),
                            'retweet' : int(retweet),
                            'likes' : int(likes),
                             })
        twitter_df = pd.DataFrame(twitter_array, columns=['tweet_id', 'retweet', 'likes'])
        
twitter_file = twitter_df.copy()
twitter_file.head(10)
```

### Act_Report
**This is the final phase of the program that shows the analysis and insight of the cleaned dataset**



**1.create a dataframe and filter out rows with ratings that are less than 20 .
This is because according to the pattern of the dataset the max rating seems to be less than 15 and the next highest is more than 30** 

```python
final_array = allfiles[(allfiles['rating_denominator']==10)&(allfiles['rating_numerator']<20)].copy()
final_array.shape
```

**2. Get the preview**

```python
final_array.head(2)
final_array.describe()
```

**3.Visualize the analysis**


```python
f,(ax1,ax2 )= plt.subplots(2,1 ,figsize=(12,9))
x = final_array['rating_numerator']
y = final_array['likes']
ax1.set(title = "Relationship between Ratings and Likes ", xlabel = "Ratings" , ylabel = "Likes")
ax1.scatter(x,y)

x = final_array['rating_numerator']
y = final_array['retweet']
ax2.set(title = "Ratings and retweets", xlabel="Ratings", ylabel="Retweets")
ax2.scatter(x,y,  c='red')
plt.show()
```

### Analysis
  
  **From the cell above we can infer that the higher Ratings a tweet gets the more likes they have it should have , as we can see that those tweets between the range of 12 and 14 have the highest number of likes and those less than 10 have an average number of likes with the likes peaking between 4 and 8 due the past standards of the ratings which were mostly under 10 but as the channel's increased the more irregular their ratings got.
    Also the same thing can be said for the retweets as more people tend to retweet a post depending on the rating given which also peaked between 12 and 14.**
